# Data Visualization

```python
import numpy as np
import pandas as pd
```

```python
#read excel file ( the file we cleanesd)
df=pd.read_excel('online_store_customer_data_cleaned.xlsx')
df.head(3)
```

```python
# Drop unwanted columns
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head(3)
```

# Line plot

```python
 dict_line={
     'year':[2016,2017,2018,2019,2020,2021],
     'price': [200,250,260,220,280,300]
 }

df_line = pd.DataFrame(dict_line)
    
```

```python
# use plot() method on the dataframe
df_line.plot('year', 'price');
```

# Bar plot

```python
#For horizontal bar:

df['Employees_status'].value_counts().plot(kind='bar');
```

```python
#For vertical bar:

df['Employees_status'].value_counts().plot(kind='barh');
```

# Pie plot

```python
df['Segment'].value_counts().plot(kind='pie');
```

# Box plot

```python
#We can plot this by one column or multiple columns. For multiple columns, we need to pass columns name in y variable as a list.

df.plot(y=['Amount_spent'],kind='box');
```

```python
import matplotlib.pyplot as plt

np.warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning) 
fig, ax = plt.subplots(figsize=(6,6)) 

df.boxplot(by = 'Employees_status', column =['Amount_spent'],ax=ax, grid = False);
```

# Histogram

```python
df.plot(y='Age',kind='hist',bins=10);
```

# KDE plot

```python
df.plot(y='Age',xlim=(0, 100),kind='kde');
```

# Scatterplot

```python
df.plot(x='Age',y='Amount_spent',kind='scatter');
```
