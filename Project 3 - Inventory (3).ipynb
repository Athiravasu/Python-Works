```python
import pandas as pd
data=pd.read_excel('Inventory.xlsx')
data
```

# Find the total stock

```python
a=data['Quantity'].sum()
a
```

# Find the total amount of the stock

```python
a=data['Amount of Products'].sum()
a
```

# Find the number of Categories

```python
a=data['Category'].value_counts()
a
```

#  Find the number of products in each category.

```python
a=data.groupby('Category').count()['Amount of Products']
a

```

# Find the number of Suppliers

```python
a=data['Supplier'].count()
a

```

# create a bar plot using the number of suppliers and modify the chart

```python
data['Supplier'].value_counts().plot(kind='bar')
```

# Create a Pie chart using the percentage values of number of categories

```python
a=data['Category'].value_counts(normalize=True)
a
```

```python
a.plot(kind='pie',autopct='1%.2f%%')
```
